```python
args = Namespace(
    processed_data_path = "./processed_data/",
    max_samples = 10,
    mode = 'a', # a : append, w : write,
)
```

```python
casehold_df = pd.read_csv('./raw_data/casehold.csv')
```

```python
contexts = []
holdings = []
binary_labels = []
relevance_labels = []


for index, row in casehold_df.iterrows():
  context = row['1']
  holding_index = row['12']

  for i in range(2,7):
    holding = row[i]
    contexts.append(context)
    holdings.append(holding)
    binary_labels.append(1 if i == int(holding_index) + 2 else 0)
    relevance_labels.append(row[i + 5] if row[i+5] != 0 else 1)

processed_df = pd.DataFrame({'context' : contexts, 'holding':holdings, 'binary_label' : binary_labels, 'relevance_label':relevance_labels})
```

```python
processed_df.to_csv(args.processed_data_path + '/casehold_processed.csv', mode=args.mode)
```
