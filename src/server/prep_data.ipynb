## EU2UK 

```python
billsum_df = pd.read_json('./../../Notebooks/processed_data/billsum_train_clean.json', lines=True)[0:500]
```

```python
billsum_df.head()
```

```python
billsum_df.iloc[0]['text']
```

```python
import re 
pattern = r"This Act may be cited as the ``(.*?)''"

def extract_text(s):
    match = re.search(pattern, s)
    return match.group(1) if match else None 
```

```python
print(extract_text(billsum_df.iloc[0]['text']))
```

```python

billsum_df['name'] = [extract_text(obj['text']) for obj in billsum_df]
```
